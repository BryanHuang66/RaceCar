```python
from source.env import RaceTrackEnv
from source.visibility import MazeConstructor
from source.reinforce_algo import SarsaLambda
```

```python
from sklearn.model_selection import ParameterGrid
import numpy as np
# 定义超参数的候选列表
env = RaceTrackEnv("env/race2.txt",5)
param_grid = {
    'num_episodes': [1000],
    'alpha': [0.05, 0.1, 0.15],
    'gamma': [0.8,0.9, 0.99],
    'lam':[0,0.3,0.5,0.7,1]
}
grid = ParameterGrid(param_grid)
results = []
for params in grid:
    algorithm = SarsaLambda(env, **params)
    algorithm.forward()
    rewards = algorithm.episode_rewards
    average_reward = np.mean(rewards)
    results.append((params, average_reward))
for params, average_reward in results:
    print("Parameters:", params)
    print("Average Reward:", average_reward)
    print()
```
