```python
import os
import numpy as np
import cv2
import numpy as np
import matplotlib.pyplot as plt
import tkinter as tk
import sys
import time
import pandas as pd
from tqdm import tqdm
import random
```

```python
from source.env import RaceTrackEnv
from source.reinforce_algo import ReinforcePolicyGradient
from source.visibility import MazeConstructor
```

```python
from sklearn.model_selection import ParameterGrid
env = RaceTrackEnv("env/race1.txt",5)
param_grid = {
    'num_episodes': [1000, 10000, 50000],
    'alpha': [0.05, 0.1, 0.15, 0.2],
    'gamma': [0.85,0.9, 0.95, 0.99]
}
grid = ParameterGrid(param_grid)
results = []
for params in grid:
    algorithm = ReinforcePolicyGradient(env, **params)
    algorithm.forward()
    rewards = algorithm.episode_rewards
    average_reward = np.mean(rewards)
    results.append((params, average_reward))
for params, average_reward in results:
    print("Parameters:", params)
    print("Average Reward:", average_reward)
    print()
```

```python
## save results
import pickle
with open('results_reinforce_race1.pkl', 'wb') as f:
    pickle.dump(results, f)
```
