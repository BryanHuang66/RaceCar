```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from collections import defaultdict
import pickle
```

```python
## save results
with open('result/results_reinforce_race1.pkl', 'rb') as f:
    result1 = pickle.load(f)
```

```python
## save results
with open('result/results_reinforce_race2.pkl', 'rb') as f:
    result2 = pickle.load(f)
```

```python
alpha_reward_list = defaultdict(list)
for item in result1:
    param, reward = item
    alpha = param['alpha']
    gamma = param['gamma']
    num_episodes = param['num_episodes']
    alpha_reward_list[f'gamma:{gamma}-episodes{num_episodes}'].append((alpha, reward))
fig, ax = plt.subplots()
data = alpha_reward_list
# 
for label, values in data.items():
    x = [v[0] for v in values]
    y = [v[1] for v in values]
    ax.plot(x, y, label=label)

# 
ax.legend(bbox_to_anchor=(1.05, 0), loc=3, borderaxespad=0)

# 
ax.set_xlabel('Alpha')
ax.set_ylabel('reward')

# 
ax.set_title('Alpha-Reward Plot')

# 
plt.show()
```
